```python
from autostorage.models import CalculationRow, CalculationGeometryLink, EnergyRow, GeometryRow
from autostorage.database import Database, select
from autostorage.types import Role

db = Database(":memory:")

calc_row = CalculationRow(program="foo", method="bar")
db.add(calc_row)

calc_row2 = CalculationRow(program="bar", method="foo")
db.add(calc_row2)

geo_row = GeometryRow(symbols=["O"], coordinates=[[1.0, 1.0, 1.0]], charge=0)  # ty:ignore[invalid-argument-type]
db.add(geo_row)

cg_link = CalculationGeometryLink(calculation_id=calc_row.id, geometry_id=geo_row.id, role=Role.input)
db.add(cg_link)

cg_link2 = CalculationGeometryLink(calculation_id=calc_row2.id, geometry_id=geo_row.id, role=Role.input)
db.add(cg_link2)
```

#### Matching directly to a model

```python
print("All calculations in database:")
stmt = select.matching_rows(CalculationRow.partial())
for row in db.exec_all(stmt):
    print(row)
```

```python
print("Calculations matching program='foo':")
stmt = select.matching_rows(CalculationRow.partial(program="foo"))
for row in db.exec_all(stmt):
    print(row)
```

```python
print("Calculations matching program='bar':")
stmt = select.matching_rows(CalculationRow.partial(program="bar"))
for row in db.exec_all(stmt):
    print(row)
```

```python
print("Calculations matching program_keywords={'foo': 'bar'}:")
stmt = select.matching_rows(CalculationRow.partial(program_keywords={'foo': 'bar'}))
for row in db.exec_all(stmt):
    print(row)
```

```python
print("All CalculationRows using input GeometryRow(symbols=['O']):")
stmt = select.linked_rows(
    row1=CalculationRow.partial(),
    row2=GeometryRow.partial(symbols=["O"]),
    link=CalculationGeometryLink.partial(role=Role.input),
    )
for row in db.exec_all(stmt):
    print(row)
```

```python
print("All CalculationRows using program='foo' and input GeometryRow(symbols=['O']):")
stmt = select.linked_rows(
    row1=CalculationRow.partial(program="foo"),
    row2=GeometryRow.partial(symbols=["O"]),
    link=CalculationGeometryLink.partial(role=Role.input),
    )
for row in db.exec_all(stmt):
    print(row)
```
